```python
import heapq

def uniform_cost_search(graph, start, goal):
    pq = [(0, start, [start])]
    visited = {}

    while pq:
        cost, node, path = heapq.heappop(pq)

        if node in visited and visited[node] <= cost:
            continue

        visited[node] = cost

        if node == goal:
            return True, cost, path

        for neighbor, edge_cost in graph.get(node, []):
            heapq.heappush(
                pq,
                (cost + edge_cost, neighbor, path + [neighbor])
            )

    return False, float("inf"), []


graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5)],
    'C': [('D', 1)],
    'D': []
}

start_node = 'A'
goal_node = 'D'

reached, cost, path = uniform_cost_search(graph, start_node, goal_node)

if reached:
    print("Goal reached")
    print("Total cost:", cost)
    print("Path:", path)
else:
    print("Goal not reached")
```
