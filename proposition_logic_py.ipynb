```python
# -------------------- SYMPY --------------------
print("\n===== PROPOSITIONAL LOGIC USING SYMPY =====")

from sympy import symbols
from sympy.logic.boolalg import And, Or, Not, Implies, Equivalent
from sympy.logic.boolalg import truth_table

# Define propositions
p, q = symbols('p q')

# Logical expressions
expr_and = And(p, q)
expr_or = Or(p, q)
expr_not = Not(p)
expr_implies = Implies(p, q)
expr_equiv = Equivalent(p, q)

# Function to print truth tables
def print_table(expr, name):
    print(f"\n{name} Truth Table:")
    for values in truth_table(expr, [p, q]):
        print(values)

# Display truth tables
print_table(expr_and, "AND")
print_table(expr_or, "OR")
print_table(expr_not, "NOT (p)")
print_table(expr_implies, "IMPLIES (p → q)")
print_table(expr_equiv, "EQUIVALENT (p ↔ q)")


# -------------------- KANREN --------------------
print("\n===== PROPOSITIONAL LOGIC USING KANREN =====")

from kanren import run, var, eq, conde

# Logical variables
a, b, r = var(), var(), var()

# AND relation
def and_rel(x, y, result):
    return conde(
        [eq(x, True),  eq(y, True),  eq(result, True)],
        [eq(x, True),  eq(y, False), eq(result, False)],
        [eq(x, False), eq(y, True),  eq(result, False)],
        [eq(x, False), eq(y, False), eq(result, False)]
    )

print("\nTruth table for AND:")
print(run(0, (a, b, r), and_rel(a, b, r)))

# OR relation
def or_rel(x, y, result):
    return conde(
        [eq(x, True),  eq(y, True),  eq(result, True)],
        [eq(x, True),  eq(y, False), eq(result, True)],
        [eq(x, False), eq(y, True),  eq(result, True)],
        [eq(x, False), eq(y, False), eq(result, False)]
    )

print("\nTruth table for OR:")
print(run(0, (a, b, r), or_rel(a, b, r)))


# -------------------- PYDATALOG --------------------
print("\n===== PROPOSITIONAL LOGIC USING PYDATALOG =====")

from pyDatalog import pyDatalog

pyDatalog.clear()

# Create logical terms
pyDatalog.create_terms('p_true, q_true, result, X')

# Facts
+ p_true(True)
+ q_true(True)

# Rule: result is true if both p and q are true
result(X) <= p_true(True) & q_true(True)

print("\nFacts:")
print(pyDatalog.ask("p_true(X)"))
print(pyDatalog.ask("q_true(X)"))

print("\nDerived result from rule (p AND q):")
print(pyDatalog.ask("result(X)"))
```
