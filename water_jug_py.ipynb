```python
from collections import deque

def water_jug_problem():
    jug1 = int(input("Enter capacity of jug1: "))
    jug2 = int(input("Enter capacity of jug2: "))
    goal = int(input("Enter goal amount to measure in jug1: "))

    start = (0, 0)
    goal_state = (goal, 0)

    queue = deque([(start, [])])
    visited = set()

    while queue:
        (x, y), path = queue.popleft()

        if (x, y) == goal_state:
            print("Solution with Actions:\n")
            print("(0, 0) -> Start")
            for action, state in path:
                print(f"{state} -> {action}")
            return

        if (x, y) in visited:
            continue

        visited.add((x, y))

        moves = [
            ((jug1, y), "Fill jug1"),
            ((x, jug2), "Fill jug2"),
            ((0, y), "Empty jug1"),
            ((x, 0), "Empty jug2"),
            # Pour jug1 -> jug2
            ((x - min(x, jug2 - y), y + min(x, jug2 - y)), "Pour jug1 to jug2"),
            # Pour jug2 -> jug1
            ((x + min(y, jug1 - x), y - min(y, jug1 - x)), "Pour jug2 to jug1"),
        ]

        for new_state, action in moves:
            if new_state not in visited:
                queue.append((new_state, path + [(action, new_state)]))

    print("No solution found.")

# Run the function
water_jug_problem()
```
